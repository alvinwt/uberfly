```python
import matplotlib as mpl
mpl.use('Agg')
%pylab inline
from django.core.management import setup_environ
from loq import settings
setup_environ(settings)
import os
os.environ['DJANGO_SETTINGS_MODULE'] = 'loq.settings'
from loq.models import Interval , Read_alignment, Library
from django.db.models import Count, Sum
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(context='talk',style='darkgrid')
from matplotlib.backends.backend_pdf import PdfPages
sns.set(palette=['#088A08','#0000FF','#FF0000','#58FA58','#2E9AFE','#FFBF00'],context='talk',style='whitegrid')
```

```python
a = Interval.objects.all().values('chr').annotate(Count('chr'))
chrs = []
data = {}
for i in a:
    data[i['chr']] = i['chr__count']
    chrs.append(i['chr'])
fig= plt.figure(figsize=(12, 10))
ax = fig.add_subplot(111)
b = pd.Series(data)
b.plot(kind='bar', ax=ax)
plt.title("Number of miRNA per Chromosome")
```

```python
"""Generates a list of dataframes for each chromosome"""
list_df=[]
list_df_cts = []
sum_df={}
for i in chrs:
    
    a = Interval.objects.filter(chr=i).order_by('start')
    libs = Library.objects.all().values_list('rescue',flat=True)
    big_df=pd.DataFrame(index=libs)
    big_df_cts=pd.DataFrame(index=libs)
    for self in a:
        reads= Read_alignment.objects.filter(intervalName=self).order_by('lib')    
        cords = np.arange(self.start-1,self.stop+1)
        w = range(self.start-1,self.stop+1)
        w = np.array([w])
 
        df = pd.DataFrame(index=cords, columns=libs, dtype='float64')
        df = df.fillna(0).convert_objects()
        df_cts = pd.DataFrame(index=cords, columns=libs).fillna(0)
        for i in reads:
            for j in range(i.start,i.stop):
                #if sum_df[self.mirName] == 0:
                #    pass
                #else:
                df[i.library.rescue][j] += float(i.normReads)
                df_cts[i.library.rescue][j] += int(i.read_counts)
        big_df[self.mirName] =df.max() 
        big_df_cts[self.mirName] =df_cts.max() 
        sum_df[self.mirName] =big_df[self.mirName].sum(axis=1)
        big_df[self.mirName] =df.max() / float(sum_df[self.mirName])*100
        #print big_df_cts
        #print sum_df
        #print big_df[self.mirName] / float(sum_df[self.mirName])*100
    
    list_df.append(np.round(big_df,2))
    list_df_cts.append(big_df_cts)
    
```

```python
"""Module to export all miRNA max values into excel spreadsheet"""
from pandas import ExcelWriter
writer = ExcelWriter('max_read_counts.xlsx')
for i, df in enumerate(list_df_cts):
    df.to_excel(writer, chrs[i])
writer.save()
#print list_df_cts[0]
#print list_df[0]
#print chrs
```
